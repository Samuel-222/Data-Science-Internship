```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
```

```python
plt.plot(x_values,y_values)
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.ylabel('some no')
plt.show()
```

```python

values=[5,6,3,7,2]
names=["a","b","c","d","e"]
a=np.array(["red","blue","green","pink","black"])
plt.bar(names,values,color=a)
plt.show()
```

```python
ages=[2,3,5,56,7,4,66,64,45,54,74,232,33,44,55,66,77,43,21,8,4,55,67,78,89]
range=(0,100)
bins=20
plt.hist(ages,bins,range,alpha=1,
color='blue',histtype='bar',edgecolor='black')

plt.xlabel("age")
plt.ylabel("people")
plt.title("hist")
```

```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.scatter(x_values,y_values,s=100,color="blue",marker="x")
plt.show()
```

```python
activities=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','y','g','b']
plt.pie(slices,labels=activities, colors=colors, startangle=90, shadow=True, explode=(0.1,0.1,0.6,0.1), radius=1.5, autopct='%1.1f%%')
plt.legend()
plt.show()
```
